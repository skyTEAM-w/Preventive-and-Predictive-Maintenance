```python
import pandas as pd

fre_data_set = pd.read_csv('../data/output/Out_40.csv', index_col=0, header=None)
fre_data_set.head()
```

```python
fre_data_set.index = pd.to_datetime(fre_data_set.index, format='%Y.%m.%d.%H.%M.%S')
fre_data_set.head()
```

```python
fre_data_set[2].plot(figsize=(20, 10))
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

scaler = MinMaxScaler()

data_pro = pd.DataFrame(scaler.fit_transform(fre_data_set.iloc[:, :101]), index=fre_data_set.index, columns=fre_data_set.columns[:101])
data_pro
```

```python
pca = PCA(n_components=10, svd_solver='full')

data_pca = pca.fit_transform(data_pro)
data_pca = pd.DataFrame(data_pca)
data_pca.index = data_pro.index

data_pca
```

```python
data_pca[:'2003-10-22 23:54:13'].plot(figsize=(20, 10))
```
